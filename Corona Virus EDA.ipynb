```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv('C:/Users/hp/Corono Virus/2019_nC0v_20200121_20200126_cleaned.csv')
data
```



    Province/State - City of virus suspected cases.
    Country - Country of virus suspected cases.
    Date last updated - Date of update of patient infected
    Confirmed - Confirmation by doctors that this patient is infected with deadly virus
    Suspected - Number of cases registered
    Recovered - Recovery of the patient
    Deaths - Death of the patient .



```python
data = data[data['Country'] == 'Mainland China']
data = data.drop(['Suspected'], axis=1)
data = data.fillna(0)
```

```python
unique_provinces = data['Province/State'].unique()
```

```python
unique_provinces =  data['Province/State'][data.Confirmed > 0].unique()
```

```python
province_confirmed_cases = []
for i in unique_provinces:
    province_confirmed_cases.append(data[data.Confirmed>0][data['Province/State']==i].Confirmed.sum())
```

### Confirm Cases againt Provinces

```python
for i in range(len(unique_provinces)):
    print(f'{unique_provinces[i]}: {province_confirmed_cases[i]} cases')
```

```python
import datetime

data['Date last updated'] = pd.to_datetime(data['Date last updated'])
data['Date'] = [datetime.datetime.date(d) for d in data['Date last updated']] 
data['Time'] = [datetime.datetime.time(d) for d in data['Date last updated']] 
```

### Confirm cases in  past 6 days

```python
world_cases = []

for i in unique_dates:
    world_cases.append(data[data['Date']==i].Confirmed.sum())

plt.figure(figsize=(15, 10))
plt.plot(unique_dates, world_cases)
plt.title('No of Coronavirus Cases Over Days', size=15)
plt.xlabel('Days', size=15)
plt.ylabel('No of Cases', size=15)
plt.show()
```

### Casualities in past 6 days

```python
world_cases = []

for i in unique_dates:
    world_cases.append(data[data['Date']==i].Deaths.sum())


plt.figure(figsize=(20, 12))
plt.plot(unique_dates, world_cases)
plt.title('No of Coronavirus Cases Over Days', size=30)
plt.xlabel('Days', size=30)
plt.ylabel('No of Deaths', size=30)
plt.show()
```

```python
province_Death_cases = []
for i in unique_provinces:
    province_Death_cases.append(data[data.Deaths>0][data['Province/State']==i].Deaths.sum())
```
